```python
import pandas as pd
import numpy as np
import missingno as msno
from matplotlib import pyplot as plt
```

```python
# 文件路径 如/home/data/your_data.csv
data_path = 'RANIAC_181210_345_sfMRI_246.csv'
# 可视化输出路径，如/home/data/show_missing.png
show_path = 'show_missing.png'
```

```python
df = pd.read_csv(data_path)
# 若文件列数大于400，则适当增大figsize的第一项
# 如列数>1000时，figsize=(64, 32)
msno.matrix(df, figsize=(64, 32), width_ratios=(24, 1))
# plt.show()
plt.savefig(show_path)
```
